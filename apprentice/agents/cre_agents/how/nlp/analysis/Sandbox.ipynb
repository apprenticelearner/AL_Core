```python
from text2math.parser import Text2MathParser
```

```python
dictionary = {
  "add" : "+",
  "sum" : "+",
  "total" : "+",
  "plus" : "+",
  "count" : "+",

  "subtract" : "-",
  "difference": "-",

  "reduce": "-",
  "minus": "-",

  "multiply" : "x", 
  "product" : "x",
  "times" : "x",
  "double" : "*2",
  
  "ratio" : "/",
  "divide" : "/",
  "quotient": "/",
  "equals" : "=",
  
  "half" : "/2",

  "ones" : "[0]",
  "ones-digit" : "[0]",
  "tens" : "[1]",
  "tens-digit" : "[1]",
}

special_patterns = {
  r"(\S+)\sdivided\sby\s(\S+)" : "/",
  r"(\S+)\stimes\s(\S+)" : "x",
  r"(\S+)\sminus\s(\S+)" : "-",
  r"(\S+)\splus\s(\S+)" : "+",

  r"ones\sdigit" : "[0]",
  r"ones'\sdigit" : "[0]",
  r"one's\sdigit" : "[0]",

  r"tens\sdigit" : "[1]",
  r"tens'\sdigit" : "[1]",
  r"ten's\sdigit" : "[1]",
}

not_main = {
    "take",
    "set",
    "find",
    "calculate",
    "apply"
}

display_options = {
    "compact" : False,
    "word_spacing" : 18,
    "distance" : 90,
    "arrow_spacing" : 12,
    "arrow_width" : 6
    #"fine_grained" : True,
}

spacy_model = 'en_core_web_trf'

parser = Text2MathParser(dictionary, special_patterns, spacy_model=spacy_model,
            display_parse=True, display_options=display_options)

import spacy
from spacy import displacy
nlp  = spacy.load(spacy_model)
def show_deps(text,display_options=display_options):
    print(display_options)
    doc = nlp(text)
    displacy.render(doc, style="dep", jupyter=True, options=display_options)
```

```python
show_deps("Divide 8 and 4", display_options={**display_options, "distance": 80, "arrow_spacing": 8})
show_deps("Divide cat and cat", display_options={**display_options, "distance": 80, "arrow_spacing": 8})
show_deps("Fetch cat and cat", display_options={**display_options, "distance": 80, "arrow_spacing": 8})
show_deps("We witches three", display_options={**display_options, "distance": 80, "arrow_spacing": 8})
```
